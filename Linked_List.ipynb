## Coding and Manipulating Linked Lists

- <b> Linked list</b> is data structure in which elements are store contiguously or linearly.
- Contrast to an array, wherein array elements are separately indexed.
<i> Disadvatanged : searching and appending in a linked list is linear O(n)</i>
<h4> Advantages/Rationale</h4>
- Insertions (prepending) and deletions are O(1) i.e. constant
- Dynamic size:  because size of arrays are fixed so deletions and insertions are expensive since you have to know the size of the array to an insertion, and elements have to be moved around to create new space when editing the array

<h4> Disadvantage</h4>
-  Only sequential access so no random access since it's organised contiguously
- need to allocate extra memory for pointer for each element.
- Can't do constant-time random access




## Singly linked list
- runs in one direction 
- a node only knows about the node after it

```python
"""Step1 implement the node type that will be stored in the linked list"""
class Node(object):
    def __init__(self,value):
        self.value = value
        self.next = None
    
    #methods for setting data field and next field pointers
    
    def get_data(self):
        return self.value
    
    def set_data(self,value):
        self.value = value
    #to make new nodes
    def get_next(self):  
        return self.next
    def set_next(self, next):
        self.next = next # where does the next come from 
```

```python
"""Step2 the actual Linked List Class """

class LinkedList(object):
    #every linkedlist needs to be initiated with a head/root pointer and size
    def __init__(self, head=None):
        self.head = head
        self.size= 0 #how many nodes
        
    def get_size(self):
        return self.size
    
    #code to print list contents
    def print_list(self):
        current_node = self.head
        while (current_node!=None):
            print("Node: ", current_node.get_data())
            current_node = current_node.get_next()
    
    #insert by prepending
    def insert(self, value):
        new_node =  Node(value)       #create a node
        new_node.set_next(self.head)  #set new node to point to head
        self.head = new_node         #update head to new node
        self.size+=1 #update size of list
        
    ##find the first node with a given value
    def find(self, target):
        current_node = self.head    #start at head of the list
        while current_node:
            if current_node.get_data() != target:
                current_node = current_node.get_next() #move one node ahead
            else:
                return current_node
        #if target value not in list
        return f"{target} not in linked list!"
        
        
    #delete node at a specified index
    def delete(self,idx):
        
        if idx > self.size-1:
            return "Index exceeds linked list size!"
        if idx ==0:
            self.head = self.head.get_next()
        else:
            counter = 0
            prev_node = self.head
            while counter < idx-1: # because we want the node preceeding the targt node
                prev_node = prev_node.get_next()
                counter+=1
            #when arrive at index preceeding target node
            prev_node.set_next(prev_node.get_next().get_next() )
            #decrement size of linked list
            self.size-= 1         
```

```python
ll = LinkedList()

for i in range(10):
    ll.insert(i)
```
